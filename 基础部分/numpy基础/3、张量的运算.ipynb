### 张量与张量的运算
加、减、乘（数乘、点乘）、除（数除）、逆、伪逆
转置

```python
a=np.random.randn(3,4)
b=np.random.randn(3,4)
c=np.random.randn(3,1) # 使用广播
a+b,a+c,a*b,a/b
```

```python
#矩阵的点乘
a=np.random.randn(3,4)
b=np.random.randn(4,5)
a@b
```

```python
a=np.random.randn(2,7,3,4)
b=np.random.randn(4,5) #广播
c=a@b
c.shape
```

```python
# 逆
a=np.random.randn(4,4)
np.linalg.inv(a)
```

```python
# 伪逆
a=np.random.randn(4,3)
np.linalg.pinv(a)
```

```python
# 转置
a=np.random.randn(3,4)
# 以下两种方式转置
a=a.transpose()
a.T
a.shape
```

```python
# 张量的转置
a=np.random.randn(3,4,5)
a=a.transpose(0,2,1) #更改维度
a.shape
```

```python
a=np.array([[1,2,3,4],[5,6,7,8]])
# 加减乘除
a+1,a-1,a*2,a/2
# 平方

a**2
# 开方
np.sqrt(a)
# 对数,以e为底、以10为底、以2为底
np.log(a),np.log10(a),np.log2(a)
# e的指数幂
np.exp(a)
```

```python
a=np.random.randn(3,5)
print (a)
# 取极值
a.max(),np.max(a)
a.min(),np.min(a)
a.mean(),np.mean(a) # 平均数
a.std(),np.std(a) # 标方差
a.var(),np.var(a)   #方差
```

```python
# 按1轴取最大值（从维度的意义上看），可以是多个维度
a.max(axis=1),np.max(a,axis=1)
# 按0轴取最大值
a.max(axis=0),np.max(a,axis=0)
```

```python
# 希望结果维度不发生变化
a.max(axis=0,keepdims=True)
```

```python
# 标准化
b=(a-a.mean(axis=1,keepdims=True))/a.std(axis=1,keepdims=True) # 不加keepdims维度会不一致
print(b)
# 等同于
# print(a.shape)
# print(a.mean(axis=1)[:,None].shape)
# print((a-a.mean(axis=1)[:,None]).shape)
# print(a.std(axis=1)[:,None].shape)
c=(a-a.mean(axis=1)[:,None])/a.std(axis=1)[:,None] 
print(c)
```

```python
# 最大值索引
a=np.random.randn(2,3,4)
print(a)
m=np.argmax(a,axis=1)
print(m)

```

## 图片


```python
from PIL import Image
import numpy as np

img = Image.open("./images/1.jpg")
# img.show()
data = np.array(img)
print(data.shape)

# data = data.reshape(2,640,2,360,3)
# data = data.transpose(0,2,1,3,4)
# data = data.reshape(4,640,360,3)

data = data.reshape(640,2,360,2,3)
data = data.transpose(1,3,0,2,4)
data = data.reshape(4,640,360,3)

img = Image.fromarray(data[0])
print(data[0].shape)
img.show()
```
